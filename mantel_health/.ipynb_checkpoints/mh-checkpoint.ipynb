```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("train.csv")
```

```python
data = data.drop('id', axis=1)
data = data.drop('Name', axis=1)
data = data.drop('City', axis=1)
```

```python
data.shape
```

```python
data.columns
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# Apply Label Encoding to each categorical column
for column in data.select_dtypes(include=['object']).columns:
    data[column] = label_encoder.fit_transform(data[column])
```

```python
data
```

```python
data.corr()["Depression"]
```

```python
data.drop(columns=["CGPA","Gender","Sleep Duration","Degree","Family History of Mental Illness"],axis=1,inplace=True)
```

```python
data = data.dropna(axis=1)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

X = data.drop('Depression', axis=1)
y = data['Depression']

acc_score = []
seeds = range(0,101)

for seed in seeds:
    print(seed)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=seed)
    log_reg = LogisticRegression()
    log_reg.fit(X_train, y_train)
    y_pred = log_reg.predict(X_test)
    logit_score = accuracy_score(y_test, y_pred)
    acc_score.append(logit_score)
```

```python
plt.plot(seeds,acc_score)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=acc_score.index(max(acc_score)))
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
y_pred = log_reg.predict(X_test)
y_pred_prob = log_reg.predict_proba(X_test)[:, 1] 
```

```python
y_pred_prob
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report, confusion_matrix

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Calculate precision, recall, F1-score, and AUC
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred_prob)

# Display classification report and confusion matrix
class_report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("AUC Score:", auc)
print("\nClassification Report:\n", class_report)
print("Confusion Matrix:\n", conf_matrix)
```
